```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp('Elon flew to mardaa yesterday. He carried biryani in his bag along with her wife.')


for token in doc:
    print(token, " | ", token.pos_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp('He quit a job.')

doc[1]
print(doc[1].text, " | ", doc[1].pos_, " | ", spacy.explain(doc[1].tag_))
```

```python
earning = """Estimate Momentum measures change in analyst sentiment over time and may be an indicator of future price movements. The Change in Consensus chart shows the current, 1 week ago, and 1 month ago consensus earnings per share (EPS*) forecasts. For the fiscal quarter endingMar 2024 , the consensus EPS* forecast has remained the same over the past week at 2.84 and increased over the past month from 2.81 to 2.84(1.07%). Of the 1 analysts making quarterly forecasts, 1raised and none lowered their forecast. For the fiscal year ending Jun 2024 , the consensus EPS* forecast has remained the same over the past week at 11.63 and increased over the past month from 11.6 to 11.63(0.26%). Of the 1 analysts making yearly forecasts, 1raised and none lowered their forecast."""
```

```python
doc = nlp(earning)
for token in doc:
    print(token, " | ", token.pos_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp(earning)
for token in doc:
    if token.pos_ in ['Space', 'X', 'PUNCT']:
        
        print(token, " | ", token.pos_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp(earning)

for token in doc:
    if token.pos_ not in ['Space', 'X', 'PUNCT']:
        
        print(token, " | ", token.pos_, " | ", spacy.explain(token.tag_))
```

```python
doc = nlp(earning)

filter_tokens = []

for token in doc:
    if token.pos_ not in ['Space', 'X', 'PUNCT']:
         filter_tokens.append(token)
```
